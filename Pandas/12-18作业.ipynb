```python
# 导入包
import numpy as np
import pandas as pd
```

## 1. 正确读取数据 

```python
data = pd.read_csv("data/practice6.csv", header=None)
data
```

## 2. 重命名列标签（改成中文或者英文，不要拼音）

```python
# 2.1 可以在读取数据之后，直接暴力执行columns
data.columns = ["类型", "年份", "里程", "地点", "售价", "原价"]
data
```

```python
# 2.2 也可以在读取数据的时候指定names
data = pd.read_csv("data/practice6.csv", names=["类型", "年份", "里程", "地点", "售价", "原价"])
data
```

## 3. 去除重复数据

```python
# 3.1 判断是否有重复值
data.duplicated().is_unique
# duplicated函数返回的布尔型Series对象值不唯一，所以存在重复值
```

```python
# 3.2 删除重复数据
data = data.drop_duplicates()
data
```

## 4. 处理缺失值

```python
# 4.1 判断是否有缺失值
data.isnull().value_counts()
# isnull的放回结果中有73处缺失值，都在原价里
```

```python
# 4.2 删除缺失值
data = data.dropna()
data
```

```python
# 删除异常数据
data = data[data["类型"] != "leixing"]
data
```

## 5. 把年份、里程、售价和原价都转成数值型

```python
data["年份"] = data["年份"].map(lambda x: np.int(x.replace("年","")))
data["里程"] = data["里程"].map(lambda x: np.float(x.replace("公里", "").replace("万", "")))
data["售价"] = data["售价"].map(lambda x: np.float(x.replace("万","").replace("补贴后","")))
data["原价"] = data["原价"].map(lambda x: np.float(x.replace("万","")))
data
```
