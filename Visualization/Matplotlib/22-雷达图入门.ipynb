````python
# 导入包
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager
````

````python
# 创建字体管理对象，指向具体的字体文件
font = font_manager.FontProperties(
    fname="/System/Library/Fonts/STHeiti Light.ttc", size=14, weight=4
)
````

## 一、使用plt.polar绘制雷达图

```python
plt.polar(x, values, ...)
```
### 参数解释
* x: 绘制雷达图数据的弧度坐标（2*PI=360 ）
* values: 数据的取值

**用Matplotlib.pyplot绘制雷达图的时候需要手动封闭，所以要在最后多添加一个起始坐标**

````python
# 创建实验数据
properties = ["臂力", "身法", "内息", "根骨", "悟性"]
# 手动实现封闭
values = [60, 85, 36, 92, 78, 60]
````

````python
# 0-360度的极坐标均分为6等份，其中0和360是重合的
theta = np.linspace(0, 2 * np.pi, 6)
# 设置绘图数据
plt.polar(theta, values)
# 填充
plt.fill(theta, values)
# 设置坐标刻度
plt.xticks(theta[0:5], properties, fontproperties=font)
# 绘图
plt.show()
````

## 二、使用子图绘制雷达图
### 2.1 plt.subplot

```python
ax = plt.subplot(111, projecto="polar")
ax.plot(x,values)
```
### 参数解释
* x: 绘制雷达图数据的弧度坐标（2*PI=360 ）
* values: 数据的取值

**用Matplotlib.pyplot绘制雷达图的时候需要手动封闭，所以要在最后多添加一个起始坐标**

````python
# 创建一个子图，设置为极坐标
ax = plt.subplot(111, projection="polar")
# 设置绘图数据
ax.plot(theta, values)
# 填充
ax.fill(theta, values)
# 绘图
plt.show()
````

### 2.2 plt.subplots

```python
figure, ax = plt.subplots(n, m, subplot_kw={"projection": "polar"})
# 如果是多个子图，即n或者m大于1
ax[i].plot(x,values)
# 如果是1个子图，即n = m = 1
ax.plot(x,values)
```
### 参数解释
* x: 绘制雷达图数据的弧度坐标（2*PI=360 ）
* values: 数据的取值

**用Matplotlib.pyplot绘制雷达图的时候需要手动封闭，所以要在最后多添加一个起始坐标**

````python
# 创建figure对象和其下的多个子图（例子中是1*1个子图），指定子图为极坐标
figure, axes = plt.subplots(1, 1, subplot_kw={"projection": "polar"})
# 设置绘图数据，如果不止一个子图，则axes是一个序列
axes.plot(theta, values)
# 填充
axes.fill(theta, values)
# 绘图
plt.show()
````

### 2.3 fig.add_subplot

```python
fig = plt.figure(...)
ax = fig.add_subplot(111, polar=True)
ax.plot(x,values)
```
### 参数解释
* x: 绘制雷达图数据的弧度坐标（2*PI=360 ）
* values: 数据的取值

**用Matplotlib.pyplot绘制雷达图的时候需要手动封闭，所以要在最后多添加一个起始坐标**

````python
# 创建figure对象
fig = plt.figure(figsize=(4, 4))
# 在figure对象中添加子图，并设置子图为极坐标
ax2 = fig.add_subplot(111, polar=True)
# 设置绘图数据
ax2.plot(theta, values)
# 填充
ax2.fill(theta, values)
# 绘图
plt.show()
````
