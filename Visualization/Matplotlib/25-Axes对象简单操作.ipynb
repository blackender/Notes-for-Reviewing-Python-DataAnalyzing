```python
import matplotlib.pyplot as plt
import numpy as np
```

## Axes对象
1. Axes对象是Matplotlib中绘制具体图形的容器
2. plt的所有的图形绘制都是对Axes对应绘图方法的封装与调用：
    * 例如，plt.plot调用的是Axes.plot；plt.scatter调用的是Axes.plot；plt.bar调用的是Axes.bar

## 一、设置坐标轴的最大值和最小值

```python
# 创建画板
fig = plt.figure()
# 在fig上添加一张子图（Axes对象）
ax1 = fig.add_subplot(111)
# 设置子图的绘图数据，绘制折线图
ax1.plot(np.random.randn(10))

# 设置x轴的最大值和最小值
ax1.set_xlim(-2, 10)
# 设置y轴的最大值和最小值
ax1.set_ylim(-3, 3)
```

## 二、添加文本

```python
# 创建画板
fig = plt.figure()
# 在fig上添加一张子图（Axes对象）
ax1 = fig.add_subplot(111)
# 设置子图的绘图数据，绘制折线图
ax1.plot(np.random.randn(10))

# 添加简单的标准文本
ax1.text(0, 0, "Hello")
```

## 三、绘制双Y轴

```python
# 创建画板
fig = plt.figure()
# 在fig上添加一张子图（Axes对象）
ax1 = fig.add_subplot(111)
# 设置子图ax1上的绘图数据，绘制柱状图
ax1.bar(np.arange(0, 10), np.random.randn(10) * 100, label="temperature")
# 设置子图ax1的y轴标签
ax1.set_ylabel("Temperature")

# 创建一个与ax1共享x轴对象的新子图ax2
ax2 = ax1.twinx()
# 设置子图ax2上的绘图数据，绘制折线图
ax2.plot(np.arange(0, 10), np.random.randn(10) * 10, color="r", label="rain")
# 设置子图ax2的y轴标签
ax2.set_ylabel("Rain")

# 统一绘制fig上所有子图的图例
fig.legend(ncol=2, loc="upper center")
```
