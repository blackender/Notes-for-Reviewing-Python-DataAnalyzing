## 作业要求

1. 读取长沙市某年5月份的气温数据（changsha.csv），按照时间顺序绘制成折线图
2. "highest"为当天最高气温,"lowest"为当天最低气温
3. 使用圆点表示数据点
4. x轴是是时间，y轴是温度，要求图中有具体刻度
5. x轴的标题是"时间（天）"，y轴的标题是”气温（摄氏度）“
6. 图标题是”长沙5月份气温走势图“

```python
# 导入通用包
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 导入字体管理包
from matplotlib import font_manager

# 创建字体管理对象，指向具体的字体文件
font = font_manager.FontProperties(
    fname="/System/Library/Fonts/STHeiti Light.ttc", size=12, weight=4
)
```

```python
# 读取数据
data = pd.read_csv("changsha.csv")
data.head()
```

### 参考代码一：

```python
# 创建画板
plt.figure(figsize=(10,4), dpi=120)
# 设置画风
plt.style.use("seaborn-bright")
# 设置绘制数据
plt.plot(data,marker="o", linestyle="--", linewidth=0.5)
# 设置表名
plt.title("长沙5月份气温走势图", font_properties=font)
# 设置x轴标签
plt.xlabel("时间（天）", font_properties=font)
# 设置y轴标签
plt.ylabel("气温（℃）", font_properties=font)
# 设置x轴刻度
plt.xticks(np.arange(31),np.arange(1,32))
# 设置注释
for column in data.columns:
    for index,value in enumerate(data[column]):
        plt.annotate(value,xy=(index,value),xytext=(index+0.1,value+0.5))

plt.show()
```

### 参考代码二：

```python
# 创建画板
plt.figure(figsize=(10,4), dpi=120)
# 设置画风
plt.style.use("bmh")
# 设置绘制数据
plt.plot(data["highest"],"o--r",linewidth=.5)
plt.plot(data["lowest"],"o--b",linewidth=.5)
# 设置表名
plt.title("长沙5月份气温走势图", font_properties=font)
# 设置x轴标签
plt.xlabel("时间（天）", font_properties=font)
# 设置y轴标签
plt.ylabel("气温（℃）", font_properties=font)
# 设置x轴刻度
plt.xticks(np.arange(31),np.arange(1,32))
# 设置注释
for column in data.columns:
    for index,value in enumerate(data[column]):
        plt.annotate(value,xy=(index,value),xytext=(index+0.1,value+0.5))

plt.show()
```
