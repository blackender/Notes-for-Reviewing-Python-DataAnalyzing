```python
# 导入包
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import font_manager
```

```python
# 创建字体管理对象，指向具体的字体文件
font = font_manager.FontProperties(
    fname="/System/Library/Fonts/STHeiti Light.ttc", size=14, weight=4
)
```

```python
# 创建实验数据
movies = {
    "长津湖之水门桥": [2.01, 4.59, 7.99, 11.83, 16],
    "这个杀手不太冷静": [3.19, 5.08, 6.73, 8.1, 9.35],
    "奇迹.笨小孩": [5.07, 6.92, 9.3, 11.29, 13.03],
    "穿过寒冬拥抱你": [2.72, 3.79, 4.45, 4.83, 5.11],
    "狙击手": [0.56, 0.74, 0.83, 0.88, 0.92],
    "四海": [0.66, 0.95, 1.1, 1.17, 1.23],
    "误杀2": [1.13, 1.96, 2.73, 3.42, 4.05],
}
movies
```

```python
# 转换成DataFrame，方便操作
movieData = pd.DataFrame(movies)
movieData
```

## 实现一

```python
# 设置画板
plt.figure(figsize=(15, 5), dpi=120)
# 设定x轴刻度的绘制坐标
xticks = np.arange(len(movies))
# 设定柱面宽度
barWidth = 0.15
# 循环获取每天数据进行绘制
i = 2
for index in movieData.index:
    # 获取一天的数据
    oneDay = movieData.iloc[index]
    # 绘制该天的全部票房数据
    plt.bar(
        xticks - i * barWidth, oneDay, width=barWidth, label="第{}天票房".format(index + 1)
    )
    i -= 1
plt.xticks(xticks, movieData.columns, font_properties=font)
# 设定图例绘制
plt.legend(prop=font)
plt.show()
```

## 实现二

```python
# 设置画板
plt.figure(figsize=(15, 5), dpi=120)
# 设定x轴刻度的绘制坐标
xticks = np.arange(len(movies))
# 设定柱面宽度
barWidth = 0.15
# 循环获取每天数据进行绘制
for index in movieData.index:
    # 获取一天的数据
    oneDay = movieData.iloc[index]
    # 绘制该天的全部票房数据
    plt.bar(
        xticks - (2 - index) * barWidth,
        oneDay,
        width=barWidth,
        label="第{}天票房".format(index + 1),
    )
plt.xticks(xticks, movieData.columns, font_properties=font)
# 设定图例绘制
plt.legend(prop=font)
plt.show()
```
