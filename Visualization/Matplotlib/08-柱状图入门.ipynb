````python
# 导入包
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from matplotlib import font_manager
````

```python
plt.bar(data, height, width, bottom, align, color)
```

* xdata: 对应x轴上的坐标，可以为列表/列表/字符串
* height(ydata): 对应y轴上的坐标，可以为列表/列表/字符串
* data是DataFrame/Dict对象，如果传递了data对象，则xdata和ydata可以为data中的列名/key
* width: 柱面的宽度
* bottom: y轴的基面开始数，默认从y=0开始
* align: 柱体的排列方式
* color: 柱面填充颜色

## 方式一

````python
# 创建实验数据
movies = {
    "长津湖之水门桥":40.62,
    "这个杀手不太冷静":26.23,
    "奇迹.笨小孩":13.77,
    "穿过寒冬拥抱你":6.65,
    "狙击手":6.01,
    "四海":5.42,
    "误杀2":2.82
}
````

````python
# 创建字体管理对象，指向具体的字体文件
font = font_manager.FontProperties(
    fname="/System/Library/Fonts/STHeiti Light.ttc", size=14, weight=4
)
````

````python
# 提取xdata 
xData = list(movies.keys())
# 提取ydata 
yData = list(movies.values())
# 设置画板
plt.figure(figsize=(15,5), dpi=120)
# 设置绘图数据
plt.bar(xData,yData, width=0.4, color="r")
# 设置x轴刻度
plt.xticks(font_properties=font)
# 设置y轴刻度
plt.yticks(np.arange(0,45,5), ["%d亿"%x for x in np.arange(0,45,5)], font_properties=font)
# 绘图
plt.show()
````

## 方式二

````python
# 创建实验数据
data = pd.DataFrame(
    {
        "电影":movies.keys(),
        "票房":movies.values()
    }
)
data
````

````python
# 设置画板
plt.figure(figsize=(15,5), dpi=120)
# 设置绘图数据
plt.bar("电影","票房",data=data, width=0.4, color="b")
# 设置x轴刻度
plt.xticks(font_properties=font)
# 设置y轴刻度
plt.yticks(np.arange(0,45,5), ["%d亿"%x for x in np.arange(0,45,5)], font_properties=font)
# 绘图
plt.show()
````
