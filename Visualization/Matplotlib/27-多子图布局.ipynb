````python
# 导入包
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

```python
fig.tight_layout(w_pad, h_pad)
```
* 默认情况下，不传参数也可以自动实现多子图布局
* w_pad: 水平间距
* h_pad: 垂直间距

```python
fig.subplot_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)
```
* 默认情况下，不传参数不会自动实现多子图布局
* left: 左边坐标
* right: 底部坐标
* right: 右边坐标
* top: 顶部坐标
* wspace: 水平间距
* hspace: 垂直间距

```python
fig.add_gridspec(rows, cols, width_ratios=None, height_ratios=None)
```
* 按照rows和cols在fig上划分格网（默认格网大小一致）
* 可以通过width_ratios设置每列格网的宽度比
* 可以通过height_ratios设置每行格网的高度比

````python
# 自定义函数：绘图
def plotMap(ax, fontsize=12):
    # 拿到传入Aexs对象，绘制折线图
    # 简化一下，只设置y轴数据绘图
    ax.plot([1, 2])
    # 设置y轴label和x轴label
    ax.set_ylabel("Y's Label", fontsize=fontsize)
    ax.set_xlabel("X's Label", fontsize=fontsize)
    # 设置Title
    ax.set_title("Title", fontsize=fontsize)
````

## 一、默认情况
默认情况下会出现多子图的元素重叠

````python
# 调用subplots方法
# 创建一个有2行2列子图的figure对象
# figure对象里有2行2列个子图，即4个子图
fig, axes = plt.subplots(2, 2)
# 产看axes元素个数
print(axes.shape)
# 循环绘制4个子图
for axline in axes:
    for ax in axline:
        plotMap(ax)
````

## 二、设置多子图布局-规则布局

### 2.1 tight_layout
通过tight_layout避免出现多子图的元素重叠

````python
# 调用subplots方法
# 创建一个有2行2列子图的figure对象
# figure对象里有2行2列个子图，即4个子图
fig, axes = plt.subplots(2, 2)
# 产看axes元素个数
print(axes.shape)
# 循环绘制4个子图
for axline in axes:
    for ax in axline:
        plotMap(ax)
# 重置layout
fig.tight_layout()
````

````python
# 调用subplots方法
# 创建一个有2行2列子图的figure对象
# figure对象里有2行2列个子图，即4个子图
fig, axes = plt.subplots(2, 2)
# 产看axes元素个数
print(axes.shape)
# 循环绘制4个子图
for axline in axes:
    for ax in axline:
        plotMap(ax)
# 重置layout,手动设置间距
# w_pad是水平间距
# h_pad是垂直间距
fig.tight_layout(w_pad=6, h_pad=4)
````

### 2.2 subplots_adjust
通过subplots_adjust避免出现多子图的元素重叠（这个不常用）

````python
# 调用subplots方法
# 创建一个有2行2列子图的figure对象
# figure对象里有2行2列个子图，即4个子图
fig, axes = plt.subplots(2, 2)
# 产看axes元素个数
print(axes.shape)
# 循环绘制4个子图
for axline in axes:
    for ax in axline:
        plotMap(ax)
# 重置layout,手动设置间距
# wspace是水平间距
# hspace是垂直间距
fig.subplots_adjust(0, 0, 1.5, 1.5, wspace=0.5, hspace=1)
````

## 三、自定义多子图布局-不规则布局
### 3.1 简单自定义布局
利用规则布局的板式，通过subplot方法逐个创建子图

````python
# 创建2行2列的第一个子图对象
plt.subplot(221)
# 创建2行2列的第三个子图对象
plt.subplot(223)
# 创建1行2列的第二个子图对象
plt.subplot(122)
````

### 3.2 复杂自定义布局

#### 3.2.1 gridspec规则格网

````python
# 创建画板
fig = plt.figure()
# 在画板上创建一个格网：例子创建3*3的格网
gs = fig.add_gridspec(3, 3)

# 添加子图对象
# 创建并添加一个子图，大小是第0行，第0列到第2列的格网
fig.add_subplot(gs[0, 0:3])
# 创建并添加一个子图，大小是第1行，第0列到第1列的格网
fig.add_subplot(gs[1, 0:2])
# 创建并添加一个子图，大小是第1行到第2行，第2列的格网
fig.add_subplot(gs[1:3, 2])
# 创建并添加一个子图，大小是第2行，第0列的格网
fig.add_subplot(gs[2, 0])
# 创建并添加一个子图，大小是第2行，第1列的格网
fig.add_subplot(gs[2, 1])
# 自动重新布局，避免子图重叠
fig.tight_layout()
````

#### 3.2.2 gridspec不规则格网

````python
# 设置宽度比
widths = (1, 2, 2)

# 设置高度比
heights = (1, 2, 1)

# 创建画板
fig = plt.figure()
# 在画板上创建一个格网：例子创建3*3的格网
gs = fig.add_gridspec(3, 3, width_ratios=widths, height_ratios=heights)

# 添加子图对象
for row in range(0, 3):
    for col in range(0, 3):
        fig.add_subplot(gs[row, col])

# 自动重新布局，避免子图重叠
fig.tight_layout()
````
