```python
# 导入包
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 读取数据
athletes = pd.read_csv("athletesALL.csv")[:4000]
athletes.head()
```

```python
# 提取数据
maleAthletes = athletes[athletes["Sex"] == "M"][["Height", "Weight"]]
maleAthletes.head()
```

```python
# 设置宽比例
widthRatios = (4, 1)
# 设置高比例
heightRatios = (1, 4)

# 创建画板
fig = plt.figure(figsize=(8, 8))
# 在画板上划分格网：2行2列
gs = fig.add_gridspec(2, 2, width_ratios=widthRatios, height_ratios=heightRatios)

# 根据格网添加第一张子图
    # 位置：第0行0列的格网
ax1 = fig.add_subplot(gs[0, 0])
# 设置子图数据，绘制直方图
ax1.hist(maleAthletes["Height"], bins=20)
# 设置子图隐藏x轴的ticklabel
ax1.xaxis.set_ticklabels([])

# 根据格网添加第二张子图
    # 位置：第1行0列的格网
ax2 = fig.add_subplot(gs[1, 0])
# 设置子图数据，绘制散点图
ax2.scatter("Height", "Weight", data=maleAthletes)
# 设置子图y轴刻度范围
ax2.set_yticks(range(0, 160, 20))
# 设置子图y轴数据显示范围
ax2.set_ylim(0, 140)

# 根据格网添加第三张子图
    # 位置：第1行1列的格网
ax3 = fig.add_subplot(gs[1, 1])
# 设置子图数据，绘制直方图，横向绘制
ax3.hist(maleAthletes["Weight"], bins=20, orientation="horizontal")
# 设置子图隐藏y轴的ticklabel
ax3.yaxis.set_ticklabels([])
# 设置子图y轴刻度范围
ax3.set_yticks(range(0, 160, 20))
# 设置子图y轴数据显示范围
ax3.set_ylim(0, 140)

# 重构子图布局
fig.tight_layout(w_pad=0.5, h_pad=0)
# 完成绘图
plt.show()
```
