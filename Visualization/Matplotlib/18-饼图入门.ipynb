````python
# 导入包
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import font_manager
````

```python
plt.pie(x, labels, explode, autopct, shadow, textprops, ...)
```
### 参数解释
* x: 每块数据在图中的占比值序列
* labels: 每个分块对应的文字序列
* explode: 设置部分数据块从饼图中分离的效果
* autopct: 设置比例信息的显示格式
* shadow: 设置阴影效果
* textprops: 设置文本显示属性

### 返回值
* patches: 图中每个分块的对象(序列)
* texts: 分块对应的名字文本对象(序列)
* autotexts: 分块对应的比例文本对象(序列)

````python
# 创建字体管理对象，指向具体的字体文件
font = font_manager.FontProperties(
    fname="/System/Library/Fonts/STHeiti Light.ttc", size=14, weight=4
)
````

````python
# 创建实验数据
phones = {
    "OPPO": 8.7,
    "Samsung": 23.4,
    "Vivo": 8.1,
    "Apple": 18.0,
    "Xiaomi": 12.7,
    "其他": 29.1,
}
````

## 一、简单绘制数据

````python
# 设置绘图数据
plt.pie(
    phones.values(),  # 传入phones的value序列作为分块占比
    labels=phones.keys(),  # 传入phones的key序列作为分块文字
    explode=(0, 0, 0, 0.1, 0, 0),  # 选择把Apple对应分块分离出来
    autopct="%.2f%%",  # 设置分块占比保留两位小数
    textprops={"fontproperties": font},  # 设定中文显示
    shadow=True,  # 设置显示阴影效果
)
# 设置图名
plt.title("Phones")
# 绘图
plt.show()
````

## 二、通过返回值调整绘制

````python
# 设置绘图数据
# 分别拿到了分块序列(patches)、分块名字文本序列(texts)和分块占比文本序列(autotexts)
patches, texts, autotexts = plt.pie(
    phones.values(),  # 传入phones的value序列作为分块占比
    labels=phones.keys(),  # 传入phones的key序列作为分块文字
    explode=(0, 0, 0, 0.1, 0, 0),  # 选择把Apple对应分块分离出来
    autopct="%.2f%%",  # 设置分块占比保留两位小数
    textprops={"fontproperties": font},  # 设定中文显示
    shadow=True,  # 设置显示阴影效果
)
# 设置图名
plt.title("Phones")
# 设置分块占比文本的显示效果
for autotext in autotexts:
    autotext.set_color("w")
    autotext.set_size(12)
# 绘图
plt.show()
````
