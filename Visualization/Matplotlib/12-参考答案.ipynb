## 作业要求

1. 读取三类学校（普通本科、中等职业教育、普通高中）2014-2018年间的招生数据（enrollment.csv），绘制柱状图。
2. x轴是年份，y轴是当年的招生人数(万人)。
3. 按年份进行分组绘制，同一年份的放在一个组。
4. 要求绘制图例，图例横向排列（提示：使用legend的ncol参数，具体用法自行查Matplotlib官网）
5. 图标题是”2014-2018普通本科、中等职业教育、普通高中招生人数“

```python
# 导入包
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 字体管理包
from matplotlib import font_manager
```

```python
# 创建字体管理对象，指向具体的字体文件
font = font_manager.FontProperties(
    fname="/System/Library/Fonts/STHeiti Light.ttc", size=14, weight=4
)
```

```python
# 读取数据
data = pd.read_csv("enrollment.csv")
data
```

## 实现一

```python
# 设置画板
plt.figure(figsize=(15, 7),dpi=120)
# 设置柱面宽度
barWidth = 0.15
# 设置x轴刻度
xticks = np.arange(2014,2019)
# 循环拿出每列数据，设置绘图数据
# 使用可以enumerate函数会自动生成序号
for cindex,col in enumerate(data.columns):
    # 一列数据，即一种学校
    one = data[col]
    # 绘制该列数据
    plt.bar(xticks-(1-cindex)*barWidth,one,width=barWidth,label=col)
    # 循环拿出该列每个数据，绘制注释
    for x,y in zip(xticks-(1-cindex)*barWidth,one):
        plt.annotate(y,xy=(x,y),xytext=(x-0.05,y+10))
# 设置表名
plt.title("2014-2018普通本科、中等职业教育、普通高中招生人数",font_properties=font)
# 设置y轴标签
plt.ylabel("招生人数（万人）",font_properties=font)
# 设置图例
plt.legend(prop=font,ncol=3)
# 设置y轴值域
plt.ylim(0,1000)
# 绘图
plt.show()
```
