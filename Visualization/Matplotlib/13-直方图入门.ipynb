````python
# 导入包
import matplotlib.pyplot as plt
````

```python
plt.hist(x, bins, range, density, cumulative, ...)
```
### 参数解释
* x: 一组数据，直方图会对这组数据进行分组
* bins: 分组信息，以数字或序列的形式出现。
    * 如果bins是数字：会按照该数字自动对数据x进行分组
    * 如果bins是序列：会按照序列定义的区间对数据x进行分组
* range，指定划分区间的最大值和最小值（用得不多）
    * 如果手动指定了bins，则无效
* density: 密度，把数量统计改成频率的密度统计
    * 频率的密度统计 = (区间的数量统计/总数据量)/区间长度
    * 默认是False
* cumulative: 统计值进行累加（用得不多）

### 返回值
* n: 每个bin中数据的个数，以数组形式出现
    * 如果设定了density，就是bin里的密度统计
* bins: bins的分组区间，以数组形式出现
* patches: 每个柱面对象，以数组形式出现

## 一、自动分组

### 1.1 基础操作

````python
# 设置画板
plt.figure(figsize=(15,5))
# 设置绘图数据
# bins为20，自动分成20组
counts,bins,patches = plt.hist(data,20,edgecolor="k")
# 查看返回数据，实际操作中可以删去
print(counts)
print(bins)
print(patches)
# 设置显示x轴刻度
plt.xticks(bins,bins)
# 绘图
plt.show()
````

### 1.2 加入频率密度统计

````python
# 设置画板
plt.figure(figsize=(15,5))
# 设置绘图数据
# bins为20，自动分成20组
# 计算频率密度统计
counts,bins,patches = plt.hist(data,20,edgecolor="k",density=True)
# 查看返回数据，实际操作中可以删去
print(counts)
print(bins)
print(patches)
# 设置显示x轴刻度
plt.xticks(bins,bins)
# 绘图
plt.show()
````

### 1.3 加入range设置

````python
# 设置画板
plt.figure(figsize=(15,5))
# 设置绘图数据
# bins为20，自动分成20组
# range设定最大值和最小值
counts,bins,patches = plt.hist(data,20,range=(0,160),edgecolor="k")
# 查看返回数据，实际操作中可以删去
print(counts)
print(bins)
print(patches)
# 设置显示x轴刻度
plt.xticks(bins,bins)
# 绘图
plt.show()
````
