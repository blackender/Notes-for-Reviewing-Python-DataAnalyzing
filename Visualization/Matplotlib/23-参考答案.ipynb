## 作业要求

1. 读取“scores.csv”中的数据。
2. 计算每个科的平均成绩。
3. 绘制所有科目平均成绩的雷达图。

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("scores.csv").drop(["num", "class"], axis=1)
data.head()
```

```python
meanScorces = data.mean()
meanScorces
```

```python
values = np.append(meanScorces, meanScorces[0])
values
```

```python
theta = np.linspace(0, 2 * np.pi, 11)
plt.polar(theta, values)
plt.fill(theta, values)
plt.xticks(theta[0:10], meanScorces.index.tolist())
plt.show()
```
