````python
# 导入包
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import font_manager
````

````python
# 创建字体管理对象，指向具体的字体文件
font = font_manager.FontProperties(
    fname="/System/Library/Fonts/STHeiti Light.ttc", size=14, weight=4
)
````

```python
plt.boxplot(x, notch, sym, vert, whis, positions, ...)
```
### 参数解释
* x: 需要绘制箱线图的数据
* notch: 是否设置置信区间，默认为False
* sym: 异常值的符号表示，默认为小圆点
* vert: 是否设置垂直绘制，默认为True
* whis: 设置上下限的系数，默认是1.5
* positions: 绘制多个箱线时，每个箱线的位置
* widths: 设置每个箱线的宽度
* labels: 每个箱线的标签
* meanline和showmeans: 是否绘制平均线，默认是False

## 一、单个箱线图的绘制

````python
# 创建实验数据
data = np.random.rand(100) * 100
data = np.append(data, np.array([-100, 180]))
data
````

````python
# 绘制箱线图
plt.boxplot(data, sym="^", labels=["Test"], meanline=True, showmeans=True)
plt.show()
````

## 二、多个箱线图的绘制

````python
# 读取实验数据
athletes = pd.read_csv("athletesAll.csv")
athletes.head()
````

````python
# 提取各国身高数据
countries = {
    "CHN": "中国",
    "JPN": "日本",
    "KOR": "韩国",
    "USA": "美国",
    "CAN": "加拿大",
    "BRA": "巴西",
    "GBR": "英国",
    "FRA": "法国",
    "ITA": "意大利",
    "ETH": "埃塞俄比亚",
    "KEN": "肯尼亚",
}

data4draw = []
for code in countries:
    acountry = athletes[athletes["NOC"] == code]["Height"].dropna()
    data4draw.append(acountry)
````

````python
# 设置画板
plt.figure(figsize=(12, 5), dpi=120)
# 设置绘图数据
plt.boxplot(data4draw)
# 设置x轴标签
plt.xticks(range(1, 12), countries.values(), fontproperties=font)
# 设置title
plt.title("各国运动员身高", fontproperties=font)
# 绘图
plt.show()
````
