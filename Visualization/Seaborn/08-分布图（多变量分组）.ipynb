```python
# 导入包
import numpy as np
import pandas as pd
import seaborn as sns
```

## 分布图：（多变量分组）

### pairplot()函数

#### 重点

1. pairplot的作用是读取多个变量绘制两两之间的分布图，基础的双变量分布图类型有：
    1. scatter（默认）：散点图
    2. kde：核密度估计
    3. hist：直方图
    4. reg：散点图的基础上做线性回归
2. vars参数设置变量的选择范围
3. kind参数设置双变量分布图类型
4. diag_kind参数这是单变量的分布图类型（斜对角上的图）

```python
# 读取数据(seaborn提供测试数据)
iris = pd.read_csv("dataset/iris.csv")
# 查看数据
iris.head()
```

```python
# 调用pairplot绘图
    # seaborn会自动绘制两个字段之间的关系散点图
    # 直方图为单个字段数据本身的分布情况
sns.pairplot(data=iris)
```

#### vars参数
通过vars参数可以指定用于绘制图形的字段

```python
# 调用pairplot绘图
    # seaborn会自动绘制两个字段之间的关系散点图
    # 直方图为单个变量本身的分布情况
    # 通过vars指定选用"sepal_length", "sepal_width", "petal_length"这三个字段的值进行绘图
sns.pairplot(data=iris, vars=["sepal_length", "sepal_width", "petal_length"])
```

#### kind参数
1. 通过kind参数可以指定双变量之间的分布图类型
    1. scatter（默认）：散点图
    2. kde：核密度估计
    3. hist：直方图
    4. reg：散点图的基础上做线性回归

```python
# 调用pairplot绘图
    # seaborn会自动绘制两个字段之间的关系散点图
    # 设定双变量分布图为线性回归
sns.pairplot(data=iris, kind="reg")
```

```python
# 调用pairplot绘图
    # seaborn会自动绘制两个字段之间的关系散点图
    # 设定双变量分布图为kde
sns.pairplot(data=iris, kind="kde")
```

```python
# 调用pairplot绘图
    # seaborn会自动绘制两个字段之间的关系散点图
    # 设定双变量分布图为直方图
sns.pairplot(data=iris, kind="hist")
```

#### diag_kind参数
1. 通过diag_kind参数可以指定单变量分布图类型
    1. auto（默认）：根据双变量的类型自动指定
    2. hist：直方图
    3. kde：核密度估计
    4. None

```python
# 调用pairplot绘图
    # seaborn会自动绘制两个字段之间的关系散点图
    # 设定单变量的分布图为kde
sns.pairplot(data=iris, diag_kind="kde")
```

```python
# 调用pairplot绘图
    # seaborn会自动绘制两个字段之间的关系散点图
    # 直方图为单个字段数据本身的分布情况
sns.pairplot(data=iris, diag_kind="hist", kind="kde")
```
