```python
# 导入包
import numpy as np
import pandas as pd
import seaborn as sns
```

## 统计图

### catplot()函数

#### 重点

1. catplot主要用于绘制分类数据图，内部封装了barplot（默认）、pointplot和countplot。
    1. barplot() ==> catplot(kind="bar")
    2. pointplot() ==> catplot(kind="point")
    3. countplot() ==> catplot(kind="count")
2. barplot图会自动进行统计（平均数、比例等），也可以使用estimator参数来修改统计函数
3. pointplot图用绘制指定数据项取值的统计变化关系
4. countplot图是用于统计指定数据项中每种取值的个数
    * 注意：只能给x和y中的一项传值

```python
# 读取数据(seaborn提供测试数据)
titanic = pd.read_csv("dataset/titanic.csv")
# 查看数据
titanic.head()
```

```python
# 绘制按“sex”取值分类的“survivied”分类统计图
    # 注意：如果使用catplot绘制bar图需要加上 kind="bar"
    # 每个bar顶部的黑色线条代表95%的置信区间
sns.catplot(x="sex", y="survived", data=titanic, kind="bar")
```

#### estimator参数
设定统计函数对y轴取值进行统计

```python
# 按sex取值分类统计总的survived值
sns.catplot(x="sex", y="survived", data=titanic, kind="bar", estimator=sum)
```

```python
# 绘制pointplot图，分别统计sex取值下survived取值的变化情况
sns.catplot(x="sex", y="survived", data=titanic, kind="point")
```

```python
# 绘制pointplot图，分别统计sex取值下survived取值的变化情况
    # 使用hue参数进一步细分组别统计
sns.catplot(x="sex", y="survived", data=titanic, kind="point", hue="class")
```

```python
# 绘制countplot图分别统计sex中取值为male和female的个数
sns.catplot(x="sex", data=titanic, kind="count")
```
