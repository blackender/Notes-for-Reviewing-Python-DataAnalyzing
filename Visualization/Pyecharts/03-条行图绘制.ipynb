```python
# 在jupyterlab中使用pyecharts的环境设置
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

```python
# 导入包
# Bar是条形图
from pyecharts.charts import Bar
from pyecharts import options as opt

# Faker是pyecharts提供的一个测试数据生成模块
from pyecharts.faker import Faker
```

## 绘制横向条形图

```python
# 链式调用创建条形图对象
bar = (
    Bar()
    .add_xaxis(Faker.choose())
    .add_yaxis("卖家1", Faker.values())
    .add_yaxis("卖家2", Faker.values())
    .reversal_axis() # 设置翻转为横向
)

# 第一次在jupyter里渲染需要加载javascript文件
bar.load_javascript()

# 渲染
bar.render_notebook()
```

## 绘制堆叠条形图

```python
# 链式调用创建条形图对象
stackBar = (
    Bar()
    .add_xaxis(Faker.choose())
    .add_yaxis("卖家1", Faker.values(), stack="stack1") # stack参数设置堆叠效果
    .add_yaxis("卖家2", Faker.values(), stack="stack1") # stack参数设置堆叠效果
)

# 渲染
stackBar.render_notebook()
```

## 绘制自定义间距条形图

```python
# 链式调用创建条形图对象
gapBar = (
    Bar()
    .add_xaxis(Faker.choose())
    .add_yaxis("卖家1", Faker.values(), category_gap="80%")
    .set_global_opts(title_opts=opt.TitleOpts(title="单柱式"))
)

# 渲染
gapBar.render_notebook()
```

```python
# 链式调用创建条形图对象
gapBar = (
    Bar()
    .add_xaxis(Faker.choose())
    .add_yaxis("卖家1", Faker.values(), gap=0)
    .add_yaxis("卖家2", Faker.values(), gap=0, category_gap="50%")
    .set_global_opts(title_opts=opt.TitleOpts(title="双柱式"))
)

# 渲染
gapBar.render_notebook()
```
