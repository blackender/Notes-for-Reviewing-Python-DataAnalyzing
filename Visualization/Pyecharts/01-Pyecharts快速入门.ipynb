```python
# 在jupyterlab中使用pyecharts的环境设置
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

### 绘制条形图

```python
# 导入pyecharts包中的Bar
from pyecharts.charts import Bar
```

```python
# 创建空的条形图对象
bar = Bar()
# 添加x轴的数据
bar.add_xaxis(["魅族20", "一加11", "摩托罗拉X40", "小米13", "OPPO Find X6", "Vivo X90"])
# 添加y轴的数据
bar.add_yaxis("店家1", [2700, 3100, 4500, 3999, 4000, 3600])

# 第一次在jupyter里渲染需要加载javascript文件
bar.load_javascript()
```

```python
# 绘图
    # 注意如果是在一般python环境中直接使用：bar.render（）
bar.render_notebook()
```

#### 链式调用

```python
(
    Bar()  # 创建条形码对象
    .add_xaxis(["魅族20", "一加11", "摩托罗拉X40", "小米13", "OPPO Find X6", "Vivo X90"]) # 添加x轴
    .add_yaxis("店家1", [2700, 3100, 4500, 3999, 4000, 3600]) # 添加y轴
    .render_notebook() # 绘图
)
```
