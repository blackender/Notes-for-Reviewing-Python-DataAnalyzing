```python
# 在jupyterlab中使用pyecharts的环境设置
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

```python
# 导入包
# Map是画地图的类
from pyecharts.charts import Map
from pyecharts import options as opt

# Faker是pyecharts提供的一个测试数据生成模块
from pyecharts.faker import Faker
```

## 世界地图

```python
world = (
    Map() # 创建Map对象
    .add("烧烤热力值", [list(z) for z in zip(Faker.country, Faker.values())], "world") # 添加数据
    .set_series_opts(label_opts=opt.LabelOpts(is_show=False)) # 设置隐藏表情
    .set_global_opts( # 设置全局配置
        title_opts=opt.TitleOpts(title="世界烧烤热力图"), # 设置图标题
        visualmap_opts=opt.VisualMapOpts(max_=200) # 设置数值可视化
    )
)

# 第一次在jupyter里渲染需要加载javascript文件
world.load_javascript()
```

```python
# 渲染
world.render_notebook()
```

## 全国地图

```python
cn = (
    Map() # 创建Map对象
    .add("烧烤热力值", [ list(z) for z in zip(Faker.provinces, Faker.values())], "china") # 添加数据
    # 设置全局配置
    .set_global_opts(
        title_opts=opt.TitleOpts(title="中国烧烤热力图"), # 设置标题
        visualmap_opts=opt.VisualMapOpts() # 按值显示不同颜色
    )
)

# 第一次在jupyter里渲染需要加载javascript文件
cn.load_javascript()
```

```python
# 渲染
cn.render_notebook()
```

## 省份地图

```python
hunancities = ["长沙市", "岳阳市", "衡阳市", "湘潭市", "株洲市", "郴州市", "永州市"]
hn = (
    Map() # 创建Map对象
    .add("烧烤热力值", [(x,y) for x,y in zip(hunancities, Faker.values())], "湖南") # 添加数据
    .set_global_opts( # 设置全局配置
        title_opts=opt.TitleOpts(title="湖南烧烤热力图"), # 设置标题
        visualmap_opts=opt.VisualMapOpts() # 设置数值可视化
    )
)
hn.load_javascript()
```

```python
# 渲染
hn.render_notebook()
```
